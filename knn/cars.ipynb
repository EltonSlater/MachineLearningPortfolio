<h1>K-NN</h1>
<p>This project uses the K-NN algorithm to predict the class of vehicles based off other atributes</p> <p>K-NN or K-nearest neighbor is a supervised learning algorithm which groups similar datapoints where k represents the number of groups and then uses these groupings to make predictions about new data</p>

```python
# Libraries
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("car.data", sep=",")
print(data.head())
```

```python
#data preprocessing
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint = le.fit_transform(list(data["maint"]))
doors = le.fit_transform(list(data["doors"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
cls = le.fit_transform(list(data["class"]))
```

```python
#x will be data used to predict
x = list (zip(buying, maint, doors, persons, lug_boot, safety))
#y will be variable that is predicted
y = list(cls)

#seperates the data into seperate training and testing data, the model is trained on x_train and y_train and then predicts y_test using x_test
x_train, x_test, y_train, y_test, = sklearn.model_selection.train_test_split(x,y, test_size=0.1)
print(x_train, y_test)
```

```python
#creates a new KNN object
model = KNeighborsClassifier(metric='euclidean', p=3,  n_neighbors=9)
#trains model
model.fit(x_train, y_train)

#prints model accuracy as a percentage in decimal form
acc = model.score(x_test, y_test)
print(acc)
```

```python
#attempts to predict y_test with X_test
y_pred = model.predict(x_test)
y_pred
```

```python
predicted = model.predict(x_test)
names = ["unnac", "acc", "good", "vgood"]
#compares the predicted class with the actual class
for x in range (len(predicted)):
    print ("predicted: ", names[predicted[x]], "Data:", x_test[x], "actual:",names[y_test[x]])
    n = model.kneighbors([x_test[x]], 9, True)
print("N:", n)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```
